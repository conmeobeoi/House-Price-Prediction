```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error as mae
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("Data/Raw.csv")
data.head()
```

```python
type_counts =  data.dtypes.value_counts()
print(type_counts)
```

```python
column_types = data.dtypes
unique_types = column_types.unique()

nan_counts = data.isna().sum()

for dtype in unique_types:
    count = 0
    columns_of_type = column_types[column_types == dtype].index
    for column in columns_of_type:
        if data[column].isna().any():
            count += 1
    print(f"{dtype}   {count}")
```

```python
#Bỏ cột id
data.drop(["Id"], axis = 1, inplace = True) 
```

```python
#Tách dữ liệu gốc ra thành kiểu dữ liệu chữ và số để tiện xử lý
numerical_column = data.select_dtypes(exclude = object).columns.tolist()[:-1]

string_column = data.select_dtypes(include='object').columns.tolist()
```

```python
#Xử lý giá trị NaN ở các cột số
for column in numerical_column:
    data[column] = data[column].fillna(data[column].mean())
```

```python
#Xử lý giá trị NaN ở cột chữ và label lại data
for column in string_column:
    data[column] = LabelEncoder().fit_transform(data[column])
```
